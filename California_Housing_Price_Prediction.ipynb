### 3. Data *Preprocessing*

```python
sample_data = pd.read_csv('/sample2.csv')
sample_data
```

```python
sample_data.columns = [i.strip() for i in sample_data.columns]
sample_data.columns
```

```python
sample_data.dropna(axis=0, inplace=True)  #axis=0 = null row will be deleted/ axis=1 =null column will be deleted, inplace=True (means to update original data set too)
```

```python
sample_data
```

```python
train.dropna(axis=0, inplace=True)
test.dropna(axis=0, inplace=True)
train
```

```python
test
```

```python
trainX = train.iloc[:, 0:-1]
#0:-1 means don't take -1 (all columns excluding -1) -1 column is label: medium_house_value, : means all columns
#: means all the columns (features columns are until median_income)
trainY = train.iloc[:,-1]
trainX
```

```python
testX = test.iloc[:, 0:-1]
#0:-1 means don't take -1 (all columns excluding -1) -1 column is label: medium_house_value, : means all columns
#: means all the columns (features columns are until median_income)
testY = test.iloc[:,-1]
testY
```

# 4. Modeling

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(trainX, trainY)
```

# 5. Evaluation

```python
ypred = model.predict(testX)
ypred #system predicted value
```

```python
testY  #this is original testing data label
```

# 6. Mean Absolute Error

```python
total =0
for i,j in zip(testY, ypred):
    #print(i,j)
    total += abs(i-j)

print("MAE", total/len(ypred))
```

# 7. Mean Square Error

```python
total = 0
for i,j in zip(testY, ypred):
  #print(i,j)
  total += (i-j)**2

print("MSE", total/len(ypred))
```
